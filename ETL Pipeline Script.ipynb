```python
import requests
import pandas as pd
from datetime import datetime
```

```python
API_KEY = "d574d6f14e5875972b2394829b447734"
```

```python
cities = [
    "Lagos",
    "Warri",
    "Ibadan",
    "Abuja",
    "Benin",
    "Port Harcourt",
    "Onitsha",
    "Enugu",
    "Zaria",
    "Jos"
]
```

```python
weather_data = []

for city in cities:

    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"

    response = requests.get(url)

    data = response.json()

    weather_data.append({
        "City Name": data["name"],
        "Temperature": data["main"]["temp"],
        "Humidity": data["main"]["humidity"],
        "Weather Condition": data["weather"][0]["description"],
        "Wind Speed": data["wind"]["speed"],
        "Date and Time": datetime.fromtimestamp(data["dt"])
    })
print(" Raw weather data successfully extracted from the API.")
```

*DATA TRANSFORMATION*

```python
# Organising the data into a structure format using pandas 
df= pd.DataFrame(weather_data)
```

```python
df.head()
```

```python
#Saving data 
df.to_csv("raw_weather_data.csv", index="False")
print("saved")
```

```python
df.info()
```

```python
#Column Rename.
df.rename(columns={
    "City Name" : "City",
    "Weather Condition" : "Weather",
    "Wind Speed": "Wind_Speed_m/s",
    "Date and Time": "Datetime"
}, inplace = True)
print("Done")
```

```python
df.head()
```

```python
#Capitalize Weather inputs
df["Weather"] = df["Weather"].str.title()
print("Done")
```

```python
df.tail()
```

```python
#Ensure Numeric Columns
df["Temperature"] = df["Temperature"].astype(float)
df["Humidity"] = df ["Humidity"]. astype(int)
df["Wind_Speed_m/s"] = df["Wind_Speed_m/s"].astype(float)
df["Datetime"] = pd.to_datetime(df["Datetime"])
print("Cleaned and transformed the weather dataset.")                      
```

```python
df.head()
```

```python
# Save cleaned dataset
df.to_csv("Cleaned_raw_weather_data.csv", index ="False")
print("Processed weather data saved successfully.")
```

*ANALYSIS*

```python
#Find the coolest city
Coolest = df.loc[df["Temperature"].idxmin()]
print(Coolest[["City","Temperature"]])
```

```python
#Find the hottest city
Hottest = df.loc[df["Temperature"].idxmax()]
print(Hottest[["City", "Temperature"]])
```

```python
#Average Temperature
Average_Temp = df["Temperature"].mean()
print(f"Average Temperature: {Average_Temp:.2f} °C")
```

```python
#Compare weather conditions
df["Weather"].value_counts()
```

```python
#Compare Temperatures across cities
df[["City", "Temperature"]].sort_values(by="Temperature", ascending=False)
```
